```python
import numpy as np  # üßÆ Biblioteca para c√°lculos num√©ricos

# üî¢ Cria√ß√£o de um array linear de -10 a 10
x = np.array(range(-10, 11))  # üèóÔ∏è Gera valores inteiros de -10 at√© 10

# üìà Plotagem b√°sica do array
plt.plot(x)  # ‚úèÔ∏è Plota os valores (√≠ndice vs valor)

# üñ•Ô∏è Exibi√ß√£o do gr√°fico
plt.show()
```

```python
# üìå Define uma fun√ß√£o simples que adiciona 1 ao valor de entrada
def f(x):
    return x + 1  # ‚ûï Retorna x + 1 (opera√ß√£o de incremento)

# üìä Calcula o resultado da fun√ß√£o para todos os valores em x
y = f(x)  # üîÑ Aplica a fun√ß√£o f a cada elemento de x (se x for array)

# üñ®Ô∏è Exibe os valores de entrada e sa√≠da
print("x = ", x)  # üìù Mostra os valores originais
print("y = f(x) = ", y)  # üí° Mostra os valores ap√≥s a transforma√ß√£o
```

```python
# üìà FUN√á√ÉO PARA CRIAR GR√ÅFICO
def escreve_grafico(x, y):
    plt.plot(x, y)        # 1Ô∏è‚É£ Plota a primeira curva (linha azul padr√£o)
    plt.plot(x, f(x), 'r+')  # 2Ô∏è‚É£ Plota a segunda curva (pontos vermelhos em '+')
    plt.show()            # üëÄ Mostra o gr√°fico na tela

# üßÆ FUN√á√ÉO MATEM√ÅTICA QUADR√ÅTICA
def f(x):
    return x**2          # üî¢ Calcula o quadrado de x (x¬≤)

# üöÄ EXECU√á√ÉO DO C√ìDIGO
escreve_grafico(x, f(x))  # üì§ Gera o gr√°fico com os resultados
```

```python
# üí≥ Array NumPy representando dados de an√°lise de cr√©dito
credito = np.array([
    # üßëüíº Cada linha representa um cliente com:
    # [idade, valor_emprestimo, status_pagamento]
    
    # üëá Dados dos clientes (15 registros)
    [75, 4600, 'positivo'],   # üëµ Cliente idoso com alto empr√©stimo, pagou bem
    [21, 850,  'negativo'],    # üë∂ Jovem com valor baixo, n√£o pagou
    [39, 840,  'negativo'],    # üë® Adulto com valor muito baixo, n√£o pagou
    [45, 150,  'negativo'],    # üë® Valor extremamente baixo, n√£o pagou
    [47, 3200, 'positivo'],    # üë© Valor alto, pagou
    [51, 3400, 'positivo'],    # üëµ Valor alto, pagou
    [47, 1200, 'negativo'],    # üë© Valor m√©dio, n√£o pagou
    [25, 1500, 'negativo'],    # üë© Jovem com valor m√©dio, n√£o pagou
    [30, 2800, 'negativo'],    # üë® Valor relativamente alto, mas n√£o pagou
    [69, 1500, 'positivo'],    # üë¥ Idoso com valor m√©dio, pagou
    [69, 1800, 'positivo'],    # üëµ Idosa com valor m√©dio-alto, pagou
    [50, 2500, 'positivo'],    # üë® Adulto maduro, pagou valor consider√°vel
    [39, 1700, 'negativo'],    # üë® Valor m√©dio, n√£o pagou
    [42, 1900, 'negativo'],    # üë© Valor m√©dio-alto, n√£o pagou
    [42, 1900, 'negativo']     # üë© Valor repetido (poss√≠vel digita√ß√£o errada de 'negativo')
])

# üîç Observa√ß√µes interessantes:
# 1. üìä Padr√£o: Clientes mais velhos (50+) tendem a ter mais 'positivo'
# 2. ‚ö†Ô∏è Exce√ß√£o: Linha 30 (30 anos, 2800) foi 'negativo' - caso at√≠pico
```

```python
# üßë‚Äçüíº Separa clientes com cr√©dito "positivo" (üëç avalia√ß√£o)
positivos = credito[credito[:,2] == "positivo"]  # ‚úÖ Filtra linhas onde coluna 2 = "positivo"

# üßë‚Äçüíº Separa clientes com cr√©dito "negativo" (üëé avalia√ß√£o)
negativos = credito[credito[:,2] == "negativo"]  # ‚ùå Filtra linhas onde coluna 2 = "negativo"

# üñºÔ∏è Cria nova figura (janela gr√°fica)
fig = plt.figure()

# üìä Adiciona √°rea de plotagem
ax1 = fig.add_subplot()

# üìç Configura√ß√µes do gr√°fico:
ax1.set_ylabel('üí∞ sal√°rio')         # ‚ÜïÔ∏è Eixo Y = sal√°rio
ax1.set_xlabel('üßì idade')           # ‚ÜîÔ∏è Eixo X = idade
ax1.set_title('üìä controle de cr√©dito')  # üè∑Ô∏è T√≠tulo do gr√°fico

# üî¥ Plota clientes negativos (cruzes vermelhas)
plt.plot(negativos[:,0].astype(int),  # üëâ Idades (convertidas para inteiro)
        negativos[:,1].astype(int),   # üëâ Sal√°rios (convertidos)
        'r+', mfc='none')             # ‚ùå Marcador: + vermelho vazio

# üí¨ Solicita dados do novo cliente:
salario = int(input('üíµ informe o salario:\n'))  # ü§ë Entrada do sal√°rio
idade = int(input('üéÇ informe a idade:\n'))     # üéà Entrada da idade

# üü¢ Plota novo cliente (c√≠rculo verde)
plt.plot(idade, salario, 'go', mfc='none')  # üë§ Marcador: ‚óã verde vazio

# üîµ Plota clientes positivos (c√≠rculos azuis)
plt.plot(positivos[:,0].astype(int),   # üëâ Idades
        positivos[:,1].astype(int),    # üëâ Sal√°rios
        'bo', mfc='none')             # ‚úÖ Marcador: ‚óã azul vazio

# üñ•Ô∏è Exibe o gr√°fico completo
plt.show()
```

```python
# üìê Importa a biblioteca matem√°tica para opera√ß√µes avan√ßadas
import math  

# üìè Fun√ß√£o que calcula a dist√¢ncia entre dois pontos no espa√ßo
def distancia_euclidiana(x, y):
    # üî¢ Converte as coordenadas de x para n√∫meros decimais
    x1 = x[0].astype(float)  # üéØ Primeira coordenada de x
    x2 = x[1].astype(float)  # üéØ Segunda coordenada de x
    
    # üéØ Extrai as coordenadas de y (assume que j√° s√£o n√∫meros decimais)
    y1 = y[0]  # üß≠ Primeira coordenada de y
    y2 = y[1]  # üß≠ Segunda coordenada de y
    
    # üßÆ C√°lculo da dist√¢ncia usando Teorema de Pit√°goras:
    # ‚àö[(diferen√ßa vertical)¬≤ + (diferen√ßa horizontal)¬≤]
    distancia = math.sqrt(pow(x2 - y2, 2) + pow(x1 - y1, 2))
    
    # üìä Exibe informa√ß√µes do c√°lculo:
    print(f'üìç valor de x = {x[0:2]}')  # Mostra coordenadas relevantes de x
    print(f'üìç valor de y = {y}')       # Mostra todas coordenadas de y
    print(f'üìè d(x,y) = {distancia:.2f}')  # Mostra dist√¢ncia com 2 casas decimais
    
    # üîô Retorna o resultado num√©rico
    return distancia
```

```python
# üì¶ Faz uma c√≥pia dos dados originais (matriz "credito")
dados = credito  

# ‚ûï Cria uma coluna extra de ZEROS para armazenar as dist√¢ncias futuras
distancias = np.zeros((dados.shape[0], 1))  

# üîÑ Adiciona essa coluna de zeros √† matriz "dados" (agora ter√° 4 colunas: idade, sal√°rio, status, dist√¢ncia)
dados = np.append(dados, distancias, axis=1)  

# üó®Ô∏è Aviso: come√ßaremos os c√°lculos!
print('Calculando as dist√¢ncias do novo objeto para todos os outros.')  

# üîÑ Para cada cliente na lista...
for i in range(0, dados.shape[0]):  
    # üìê Calcula a dist√¢ncia entre o cliente atual (dados[i,:]) e o novo ponto (idade, salario)
    #    e guarda o resultado na 4¬™ coluna (dados[i, 3])
    dados[i, 3] = distancia_euclidiana(dados[i, :], [idade, salario])  

# üî¢ Garante que a coluna de dist√¢ncias √© do tipo float (n√∫meros decimais)
dados[:, 3] = dados[:, 3].astype(float)  

# üîÑ Ordena TODOS os clientes com base na coluna de dist√¢ncias (do mais pr√≥ximo ao mais distante)
dados = dados[dados[:, 3].astype(float).argsort(kind='mergesort')]  

# üó®Ô∏è Aviso: mostraremos os dados ordenados!
print('Conjunto de dados ordenados pela dist√¢ncia:')  

# üñ®Ô∏è Imprime a matriz final, agora ordenada pela dist√¢ncia
print(dados)  
```

```python
# üß† Fun√ß√£o KNN (K-Nearest Neighbors) - Classificador simples baseado em vizinhan√ßa
def knn(array_ordenado, n_vizinhos, indice_classe):
    resultado = []  # üì¶ Lista vazia para armazenar classes dos vizinhos
    
    # üîç Loop que analisa os 'n_vizinhos' mais pr√≥ximos (ex: 3 vizinhos)
    for i in range(1, n_vizinhos + 1):  # ‚û° Pega do 1¬∫ ao n-√©simo vizinho
        resultado.append(array_ordenado[i, indice_classe])  #Adiciona a classe do vizinho
        
    # üèÜ Retorna a classe mais frequente (moda) entre os vizinhos
    return max(resultado, key=resultado.count)  

# üöÄ Execu√ß√£o do KNN
print('üîç Executando KNN...')
print(f'üéØ Classifica√ß√£o prevista: {knn(dados, 3, 2)}')  # üìä Usa 3 vizinhos e classe na coluna 2
```
