```python
import numpy as np  # 🧮 Biblioteca para cálculos numéricos

# 🔢 Criação de um array linear de -10 a 10
x = np.array(range(-10, 11))  # 🏗️ Gera valores inteiros de -10 até 10

# 📈 Plotagem básica do array
plt.plot(x)  # ✏️ Plota os valores (índice vs valor)

# 🖥️ Exibição do gráfico
plt.show()
```

```python
# 📌 Define uma função simples que adiciona 1 ao valor de entrada
def f(x):
    return x + 1  # ➕ Retorna x + 1 (operação de incremento)

# 📊 Calcula o resultado da função para todos os valores em x
y = f(x)  # 🔄 Aplica a função f a cada elemento de x (se x for array)

# 🖨️ Exibe os valores de entrada e saída
print("x = ", x)  # 📝 Mostra os valores originais
print("y = f(x) = ", y)  # 💡 Mostra os valores após a transformação
```

```python
# 📈 FUNÇÃO PARA CRIAR GRÁFICO
def escreve_grafico(x, y):
    plt.plot(x, y)        # 1️⃣ Plota a primeira curva (linha azul padrão)
    plt.plot(x, f(x), 'r+')  # 2️⃣ Plota a segunda curva (pontos vermelhos em '+')
    plt.show()            # 👀 Mostra o gráfico na tela

# 🧮 FUNÇÃO MATEMÁTICA QUADRÁTICA
def f(x):
    return x**2          # 🔢 Calcula o quadrado de x (x²)

# 🚀 EXECUÇÃO DO CÓDIGO
escreve_grafico(x, f(x))  # 📤 Gera o gráfico com os resultados
```

```python
# 💳 Array NumPy representando dados de análise de crédito
credito = np.array([
    # 🧑💼 Cada linha representa um cliente com:
    # [idade, valor_emprestimo, status_pagamento]
    
    # 👇 Dados dos clientes (15 registros)
    [75, 4600, 'positivo'],   # 👵 Cliente idoso com alto empréstimo, pagou bem
    [21, 850,  'negativo'],    # 👶 Jovem com valor baixo, não pagou
    [39, 840,  'negativo'],    # 👨 Adulto com valor muito baixo, não pagou
    [45, 150,  'negativo'],    # 👨 Valor extremamente baixo, não pagou
    [47, 3200, 'positivo'],    # 👩 Valor alto, pagou
    [51, 3400, 'positivo'],    # 👵 Valor alto, pagou
    [47, 1200, 'negativo'],    # 👩 Valor médio, não pagou
    [25, 1500, 'negativo'],    # 👩 Jovem com valor médio, não pagou
    [30, 2800, 'negativo'],    # 👨 Valor relativamente alto, mas não pagou
    [69, 1500, 'positivo'],    # 👴 Idoso com valor médio, pagou
    [69, 1800, 'positivo'],    # 👵 Idosa com valor médio-alto, pagou
    [50, 2500, 'positivo'],    # 👨 Adulto maduro, pagou valor considerável
    [39, 1700, 'negativo'],    # 👨 Valor médio, não pagou
    [42, 1900, 'negativo'],    # 👩 Valor médio-alto, não pagou
    [42, 1900, 'negativo']     # 👩 Valor repetido (possível digitação errada de 'negativo')
])

# 🔍 Observações interessantes:
# 1. 📊 Padrão: Clientes mais velhos (50+) tendem a ter mais 'positivo'
# 2. ⚠️ Exceção: Linha 30 (30 anos, 2800) foi 'negativo' - caso atípico
```

```python
# 🧑‍💼 Separa clientes com crédito "positivo" (👍 avaliação)
positivos = credito[credito[:,2] == "positivo"]  # ✅ Filtra linhas onde coluna 2 = "positivo"

# 🧑‍💼 Separa clientes com crédito "negativo" (👎 avaliação)
negativos = credito[credito[:,2] == "negativo"]  # ❌ Filtra linhas onde coluna 2 = "negativo"

# 🖼️ Cria nova figura (janela gráfica)
fig = plt.figure()

# 📊 Adiciona área de plotagem
ax1 = fig.add_subplot()

# 📍 Configurações do gráfico:
ax1.set_ylabel('💰 salário')         # ↕️ Eixo Y = salário
ax1.set_xlabel('🧓 idade')           # ↔️ Eixo X = idade
ax1.set_title('📊 controle de crédito')  # 🏷️ Título do gráfico

# 🔴 Plota clientes negativos (cruzes vermelhas)
plt.plot(negativos[:,0].astype(int),  # 👉 Idades (convertidas para inteiro)
        negativos[:,1].astype(int),   # 👉 Salários (convertidos)
        'r+', mfc='none')             # ❌ Marcador: + vermelho vazio

# 💬 Solicita dados do novo cliente:
salario = int(input('💵 informe o salario:\n'))  # 🤑 Entrada do salário
idade = int(input('🎂 informe a idade:\n'))     # 🎈 Entrada da idade

# 🟢 Plota novo cliente (círculo verde)
plt.plot(idade, salario, 'go', mfc='none')  # 👤 Marcador: ○ verde vazio

# 🔵 Plota clientes positivos (círculos azuis)
plt.plot(positivos[:,0].astype(int),   # 👉 Idades
        positivos[:,1].astype(int),    # 👉 Salários
        'bo', mfc='none')             # ✅ Marcador: ○ azul vazio

# 🖥️ Exibe o gráfico completo
plt.show()
```

```python
# 📐 Importa a biblioteca matemática para operações avançadas
import math  

# 📏 Função que calcula a distância entre dois pontos no espaço
def distancia_euclidiana(x, y):
    # 🔢 Converte as coordenadas de x para números decimais
    x1 = x[0].astype(float)  # 🎯 Primeira coordenada de x
    x2 = x[1].astype(float)  # 🎯 Segunda coordenada de x
    
    # 🎯 Extrai as coordenadas de y (assume que já são números decimais)
    y1 = y[0]  # 🧭 Primeira coordenada de y
    y2 = y[1]  # 🧭 Segunda coordenada de y
    
    # 🧮 Cálculo da distância usando Teorema de Pitágoras:
    # √[(diferença vertical)² + (diferença horizontal)²]
    distancia = math.sqrt(pow(x2 - y2, 2) + pow(x1 - y1, 2))
    
    # 📊 Exibe informações do cálculo:
    print(f'📍 valor de x = {x[0:2]}')  # Mostra coordenadas relevantes de x
    print(f'📍 valor de y = {y}')       # Mostra todas coordenadas de y
    print(f'📏 d(x,y) = {distancia:.2f}')  # Mostra distância com 2 casas decimais
    
    # 🔙 Retorna o resultado numérico
    return distancia
```

```python
# 📦 Faz uma cópia dos dados originais (matriz "credito")
dados = credito  

# ➕ Cria uma coluna extra de ZEROS para armazenar as distâncias futuras
distancias = np.zeros((dados.shape[0], 1))  

# 🔄 Adiciona essa coluna de zeros à matriz "dados" (agora terá 4 colunas: idade, salário, status, distância)
dados = np.append(dados, distancias, axis=1)  

# 🗨️ Aviso: começaremos os cálculos!
print('Calculando as distâncias do novo objeto para todos os outros.')  

# 🔄 Para cada cliente na lista...
for i in range(0, dados.shape[0]):  
    # 📐 Calcula a distância entre o cliente atual (dados[i,:]) e o novo ponto (idade, salario)
    #    e guarda o resultado na 4ª coluna (dados[i, 3])
    dados[i, 3] = distancia_euclidiana(dados[i, :], [idade, salario])  

# 🔢 Garante que a coluna de distâncias é do tipo float (números decimais)
dados[:, 3] = dados[:, 3].astype(float)  

# 🔄 Ordena TODOS os clientes com base na coluna de distâncias (do mais próximo ao mais distante)
dados = dados[dados[:, 3].astype(float).argsort(kind='mergesort')]  

# 🗨️ Aviso: mostraremos os dados ordenados!
print('Conjunto de dados ordenados pela distância:')  

# 🖨️ Imprime a matriz final, agora ordenada pela distância
print(dados)  
```

```python
# 🧠 Função KNN (K-Nearest Neighbors) - Classificador simples baseado em vizinhança
def knn(array_ordenado, n_vizinhos, indice_classe):
    resultado = []  # 📦 Lista vazia para armazenar classes dos vizinhos
    
    # 🔍 Loop que analisa os 'n_vizinhos' mais próximos (ex: 3 vizinhos)
    for i in range(1, n_vizinhos + 1):  # ➡ Pega do 1º ao n-ésimo vizinho
        resultado.append(array_ordenado[i, indice_classe])  #Adiciona a classe do vizinho
        
    # 🏆 Retorna a classe mais frequente (moda) entre os vizinhos
    return max(resultado, key=resultado.count)  

# 🚀 Execução do KNN
print('🔍 Executando KNN...')
print(f'🎯 Classificação prevista: {knn(dados, 3, 2)}')  # 📊 Usa 3 vizinhos e classe na coluna 2
```
